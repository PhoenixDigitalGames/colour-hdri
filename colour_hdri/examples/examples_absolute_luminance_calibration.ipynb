# Colour - HDRI - Examples: Absolute Luminance Calibration

Through this example, a radiance image courtesy of Unity Technologies will be calibrated for absolute *Luminance* using the method described in Lagarde (2016).

```python
%matplotlib inline
```

```python
from __future__ import division

import numpy as np
import os
from pprint import pprint

import colour
from colour.plotting import *

from colour_hdri import (
    EXAMPLES_RESOURCES_DIRECTORY,
    absolute_luminance_calibration_Lagarde2016,
    upper_hemisphere_illuminance_weights_Lagarde2016)

colour.utilities.filter_warnings(True, False)

colour_plotting_defaults()

OETF = colour.OETFS['sRGB']

RESOURCES_DIRECTORY = os.path.join(EXAMPLES_RESOURCES_DIRECTORY,
                                   'unity_001')
```

## Absolute Luminance Calibration

```python
HDRI_IMAGE = colour.read_image(
    os.path.join(RESOURCES_DIRECTORY, 
                 'Unity_Treasure_Island_White_Balanced.exr'))

image_plot(OETF(HDRI_IMAGE), 
           'Unity - Treasure Island')

image_plot(OETF(absolute_luminance_calibration_Lagarde2016(HDRI_IMAGE, 51000)), 
           'Unity - Treasure Island - Calibrated',
           label_colour='black')
```

## Upper Hemisphere Illuminance Weights

```python
image_plot(OETF(colour.utilities.normalise_maximum(
                    upper_hemisphere_illuminance_weights_Lagarde2016(256, 512))), 
           'Upper Hemisphere Illuminance Weights - Normalised')
```
